# Generación de paleta de colores a partir de imágenes.

### 1. Exploración y perfilamiento de los datos

#### 1.1 Se importan las librerías necesarias
Se importan las librerías `Pandas`,`Numpy` `scikit-learn` y `matplotlib`.

```python
import matplotlib.pyplot as plt
import os
import cv2
import gc
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans, MeanShift, estimate_bandwidth
```

#### 1.2 Carga de la imagen
La función `load_image` es la encargada de cargar la imagen, redimencionarla a una escala de 200x200 pixeles para que el proceso no sea tan pesado de ejecutar.

```python
def load_image(filename):
    img_path = os.path.join(folder_path, filename)
    image = cv2.imread(img_path)
    image = cv2.resize(image, (200, 200))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

#### 1.3 Preparación y normalización de la imagen

La función `prepare_image` convertierte la imagen a una matriz de píxeles (N, 3), igualmente normalizamos la información de los pixeles para trabajar en el formato [0 - 1].

```python
def prepare_image(img):
    data = img.reshape(-1, 3)
    data = data / 255.0
    return data
```

#### 1.4 Reducción de dimensionalidad

Finalmente, se usa el algoritmo de PCA para reducir la dimensionalidad de la imagen, pasando de 3D a 2D.

```python
def apply_psa(data_in_pixels):
    pca = PCA(n_components=2)
    x_reduced = pca.fit_transform(data_in_pixels)
    return pca, x_reduced
```

### 2. Desarrollo de los modelos de agrupación para identificar los colores presentes en las imágenes

Para identifica los clores presentes en las imágenes se entrenarán dos modelos no supervisados: K-means y Meanshift; para los cuales se realizará la búsqueda de los respectivos hiperparámetros que mejor se adecuan a cada imagen. Para este ejericio hemos optado por el algoritmo K-means debido a que generp mejores resultado y es más facil de controlar, esta desición la explicaremos a fondo en las conclusiones.

#### 2.1 Algoritmo K-means

K-Means es un algoritmo de clustering que agrupa los píxeles en un número específico de clusters, en este caso colores representativos. Por lo cual, el hiperparámetro a encontrar es es mejor número de cluster. Para ello, definimos la función `optimal_kmeans` para encontrar el mejor número de clusters mediante el método de la silueta con `silhouette_score`.

```python
def optimal_kmeans(x_reduced, k_min=2, k_max=5):
    # Rango de valores para n_clusters
    n_clusters_range = range(k_min, k_max)
    
    # Almacenar resultados
    resultados_kmeans = []
    
    # Búsqueda de hiperparámetros
    for n_clusters in n_clusters_range:
        # Aplicar K-Means en el espacio reducido
        kmeans = KMeans(init='k-means++',max_iter=100,n_clusters=n_clusters, random_state=42)
        kmeans.fit(x_reduced)
        
        # Calcular el silhouette score
        silhouette = silhouette_score(x_reduced, kmeans.labels_)
        resultados_kmeans.append((n_clusters, silhouette))
    return resultados_kmeans
```

#### 2.2 Obtener el mejor número de clusters 

Utilizando la función `optimal_kmeans` con un rango desde 4 hasta 7, se gráfican los valores obtenidos por el método de la silueta.

Posteriormente, se selecciona el mejor valor de `n_clusters` como el valor máximo en el conjunto obtenido con la función `optimal_kmeans`.

```python
def get_better_cluster(x_reduced):
    kmeans_df = pd.DataFrame(optimal_kmeans(x_reduced, 4, 7), columns=["n_clusters", "silhouette_score"])
    best_n_clusters = kmeans_df.loc[kmeans_df["silhouette_score"].idxmax(), "n_clusters"]
    return best_n_clusters
```

#### 2.3 Entrenamiento del modelo.

Conociendo el mejor `n_clusters`, se entrena el modelo final de K-means para obtener la paleta de color de la imagen. Para esto, de debe invertir PCA utilizando `pca.inverse_transform`.

```python
def train_model(best_n_clusters,x_reduced,pca):
    # Entrenar el modelo final con el mejor n_clusters
    kmeans_final = KMeans(n_clusters=best_n_clusters, random_state=77)
    kmeans_final.fit(x_reduced)
    
    # Obtener los centroides en el espacio reducido
    centroides_reducidos = kmeans_final.cluster_centers_
    
    # Recuperar los colores originales usando PCA inversa
    centroides_rgb = pca.inverse_transform(centroides_reducidos)
    
    # Los valores deben estar en el rango [0, 1]
    centroides_rgb = np.clip(centroides_rgb, 0, 1)
    return centroides_rgb
```

#### 2.4 Visualización de resultados

La función `plot_result` es la encargada de mostrar, la imagen de colores, la ejecución del algoritmo de PCA y la paleta de colores generada por el algortimo.

```python
def plot_result(img, x_reduced, data_in_pixels, centroids, mejor_n_clusters):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Imagen original
    axes[0].imshow(img)
    axes[0].set_title("Imagen Original")
    axes[0].axis("off")

    # Visualización con PCA
    axes[1].scatter(x_reduced[:, 0], x_reduced[:, 1], s=1, c=data_in_pixels, alpha=0.5)
    axes[1].set_title("Visualización de colores con PCA")
    axes[1].set_xlabel("Componente Principal 1")
    axes[1].set_ylabel("Componente Principal 2")

    # Paleta de colores K-means
    for i, color in enumerate(centroids):
        axes[2].fill_between([i, i+1], 0, 1, color=color)
    axes[2].set_xlim(0, mejor_n_clusters)
    axes[2].axis("off")
    axes[2].set_title(f"Paleta de colores K-means ({mejor_n_clusters} colores)")

    plt.tight_layout()
    plt.show()
```

### 3. Ejecución del programa

Finalmente se agrupan las funciones creadas en 1 sola para facilidad de ejecución del pipeline.

```python
def process_image(filename):
    img = load_image(filename)
    data_in_pixels = prepare_image(img)
    pca_model, x_reduced = apply_psa(data_in_pixels)
    best_n_clusters = get_better_cluster(x_reduced)
    centroids = train_model(best_n_clusters,x_reduced, pca_model)
    plot_result(img, x_reduced,data_in_pixels,centroids,best_n_clusters)
    collected = gc.collect()

# Ruta de la carpeta con imágenes
folder_path = "./images"
```
